# import pandas

```python
import pandas as pd
```

# connect to exel

```python
df = pd.read_excel(r"cleaning_sample.xlsx")
df
```

## delete duplicates row

```python
df = df.drop_duplicates()
df
```

## drop the not emportent column

```python
df.drop( columns="number")
```

# str.strip()

### if strip() is empty just delete the whit space in right and left text

```python
df["last_name"].str.strip()
```

### if strip("any") writ text between the parentheses its delete the sample text in your text

```python
df["last_name"].str.strip("_")
```

### rstrip() and lstrip() its delete the just right or just left sample text in your text

```python
df["last_name"].str.rstrip("/")
```

### you can use the strip("._/") its means delete . or _ or / in your text

```python
df["last_name"] = df["last_name"].str.strip("._/")
df["last_name"]
```

### tip: pay attention just chang column you need. its means use df["target column"] = df["target column"].str.strip()

### tip2: if you have a bug in between the text you cant use strip you can use re library in python or use str.replace() in pandas library

# str.replace()

### str.replace() this function can replace target text to your text.
##### for input one use your target text and for input two use thing you replace that.

```python
df["last_name"] = df["last_name"].str.replace('/', '')
df["last_name"]
```

### if you use the replace your column type must str so for phon number must chang type int to str.

```python
df["phon"] = df["phon"].apply(lambda x: str(x))
df
```

### in top line we get one row in rows and with lambda chang type to string and we used apply to set output lambda to one row in phon column.

##### tip3: Suppose that lambda is a mini function in python

##### tip4: apply function set your data for ech case.

```python
df["phon"] = df["phon"].str.replace('_', '')
df["phon"] = df["phon"].str.replace('@', '')
df["phon"] = df["phon"].str.replace('.', '')
df["phon"]
```

### and now we can shap the phon number to -> 123-456-7890

```python
df["phon"] = df["phon"].apply(lambda x: x[:3] + '-' + x[3:6] + '-' + x[6:])
df["phon"]
```

# str.split()
### sometimes we need to segment our text.
### we can use split function to segment data.
### for example we need the first three number in phon_number number.

```python
df["phon"].str.split("-", expand=True)
```

##### tip5: expand value for ech "-" make one column
### we can get index to target column. for example: 

```python
df["phon"].str.split("-", expand=True)[0]
```

### ok and now cleaning  the other column like the above columns.

```python
df
```

```python
df["phon"] = df["phon"].str.replace("nan--", "without data")
df["gender"] = df["gender"].str.replace("male", 'm')
df["gender"] = df["gender"].str.replace("fm", 'f')
df["gender"] = df["gender"].str.replace("m", 'male')
df["gender"] = df["gender"].str.replace("f", 'female')
df
```

### format data is good but the some row not have data.
### We have to decide what to do with that row.
### in this case if row not have:
##### 1) name -> delete row.
##### 2) phon number -> delete row.
##### 3) are you ok -> set not ok for this person.

# fillna()
### with this function chang Nan data to invalid data.

```python
df = df.fillna('without data')
df
```

# delete row 

```python
for x in df.index:
    if df.loc[x, "name"] == "without data" or df.loc[x, "phon"] == "without data":
        df.drop(x, inplace=True)
df
```

##### tip6: inplace value bydefalt its false but if its true dont need to write the df = df.drop(x).

##### tip7: another way to delete null values is dropna(subset="target column")

```python
df["are_you_ok"] = df["are_you_ok"].str.replace("without data", "not ok")
df
```

### your data its cleaned but we drop some row and index numbers is out of order.

# reset_index()

```python
df = df.reset_index(drop=True)
df
```
